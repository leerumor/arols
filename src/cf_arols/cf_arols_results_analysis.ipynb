```python
import pandas as pd
param_name = 'minConf'
results = pd.read_csv('./cf_validation_results/birch_cf_arols_'+param_name+'.csv')
```

```python
import numpy as np

#统计均值
for param in results[results.cluster==0].param:
    results.loc[len(results)] = [6, param, np.mean(results[results.param==param]['n_rules'])
                                , np.mean(results[results.param==param]['precisionCF'])
                                , np.mean(results[results.param==param]['recallCF'])
                                , np.mean(results[results.param==param]['f1scoreCF'])
                                , np.mean(results[results.param==param]['precisionComb'])
                                , np.mean(results[results.param==param]['recallComb'])
                                , np.mean(results[results.param==param]['f1scoreComb'])]
```

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
t = results[results.cluster==0].param

plt.figure(figsize=(10, 6))
ax1 = plt.subplot(311)
plt.plot(t, results[results.cluster==6]['precisionCF'], '-g^')
plt.plot(t, results[results.cluster==6]['precisionComb'], '-rs')
#plt.setp(ax1.get_xticklabels(), fontsize=6)
#plt.ylabel('%')
plt.setp(ax1.get_xticklabels(), visible=False)
plt.ylim(0, 40)
plt.ylabel('Precision(%)')

# share x only
ax2 = plt.subplot(312, sharex=ax1)
plt.plot(t, results[results.cluster==6]['recallCF'], '-g^')
plt.plot(t, results[results.cluster==6]['recallComb'], '-rs')
plt.ylabel('Recall(%)')

# make these tick labels invisible
plt.setp(ax2.get_xticklabels(), visible=False)
plt.ylim(0, 40)
# share x and y
ax3 = plt.subplot(313, sharex=ax1, sharey=ax2)
plt.plot(t, results[results.cluster==6]['f1scoreCF'], '-g^', label='Clustering+ItemCF')
plt.plot(t, results[results.cluster==6]['f1scoreComb'], '-rs', label='CF-AROLS')
plt.ylabel('F1 score(%)')
plt.xlim(0.8, 0.98)
plt.ylim(0, 40)
plt.xlabel(param_name)
plt.legend()
plt.savefig('./images/birch_cf_arols_'+ param_name + '.png')
plt.show()
```
