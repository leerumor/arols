```python
import numpy as np

path = './nlp_validate/birch6/'

records = dict()
vec_list = [128, 256, 400, 512]

for vector_size in vec_list:
    records[vector_size] = np.load(path + 'lstm_word2vec_dim_'+str(vector_size)+'.npy').item()
```

```python
def avg(records, vector_size, param_name):
    tmp = list()
    for cluster in range(0, 6):
        if cluster == 2: continue
        tmp.append(records[vector_size][cluster][param_name])
    results = list()
    tmp = np.array(tmp)
    length = tmp.shape[1]
    for i in range(0, length):
        results.append(np.mean(tmp[:, i].reshape(1, 5)))
    return results
```

```python
import matplotlib.pyplot as plt

# 画一个四个格子的图
np.random.seed(42)
cluster = 0

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(6, 6))
vector_size = 128
axs[0, 0].plot(avg(records, vector_size, 'loss'), label='train')
axs[0, 0].plot(avg(records, vector_size, 'val_loss'), label='test')
#axs[0, 0].set_ylabel('Cosine proximity')
#axs[0, 0].set_xlabel('Epochs')
axs[0, 0].set_title('vector_size='+str(vector_size))
axs[0, 0].legend()

vector_size = 256
axs[0, 1].plot(avg(records, vector_size, 'loss'), label='train')
axs[0, 1].plot(avg(records, vector_size, 'val_loss'), label='test')
#axs[0, 1].set_ylabel('Cosine proximity')
#axs[0, 1].set_xlabel('Epochs')
axs[0, 1].set_title('vector_size='+str(vector_size))
axs[0, 1].legend()

vector_size = 400
axs[1, 0].plot(avg(records, vector_size, 'loss'), label='train')
axs[1, 0].plot(avg(records, vector_size, 'val_loss'), label='test')
#axs[1, 0].set_ylabel('Cosine proximity')
#axs[1, 0].set_xlabel('Epochs')
axs[1, 0].set_title('vector_size='+str(vector_size))
axs[1, 0].legend()

vector_size = 512
axs[1, 1].plot(avg(records, vector_size, 'loss'), label='train')
axs[1, 1].plot(avg(records, vector_size, 'val_loss'), label='test')
#axs[1, 1].set_ylabel('Cosine proximity')
#axs[1, 1].set_xlabel('Epochs')
axs[1, 1].set_title('vector_size='+str(vector_size))
axs[1, 1].legend()
fig.text(0.57, 0.01, 'Epochs', ha='center')
fig.text(0.01, 0.5, 'Cosine proximity', va='center', rotation='vertical')
fig.tight_layout()
plt.savefig('./images/birch_lstm_arols_vec_size_avg.png')
plt.show()
```
