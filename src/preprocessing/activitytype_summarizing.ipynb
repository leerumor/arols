```python
import numpy as np
import pandas as pd
from pandas.api.types import CategoricalDtype
import os
student_learn= pd.read_csv('./data/student_train2.csv', 
                           dtype={'id_student':np.uint32, 'id_site': np.uint32, 'date': np.int16, 'sum_click':np.uint8})
#student_log = student_log.set_index()
student_learn.drop(columns=['Unnamed: 0'], inplace=True)
student_learn.head()
```

```python
student_stat = pd.read_csv('./data/studentstat.csv', dtype={'id_student':np.uint32, 'n_records': np.uint16})
student_learn['activity_type'] = student_learn['activity_type'].astype('category')
student_learn.info()
```

```python
student_learn['activity_type'].cat.categories
```

```python
activity_count = pd.DataFrame(0,columns = student_learn['activity_type'].cat.categories, index=student_stat['id_student'])
```

```python
activity_count.head()
```

```python
for row_index,row in student_learn.iterrows():
    activity_count.loc[row['id_student'],row['activity_type']] += row['sum_click']
```

```python
activity_count.head()
```

```python
activity_count.to_csv('./data/click_count.csv')
```

```python
page_count = pd.DataFrame(0,columns = student_learn['activity_type'].cat.categories, index=student_stat['id_student'])
```

```python
page_count.head()
```

```python
for row_index,row in student_learn.iterrows():
    page_count.loc[row['id_student'],row['activity_type']] += 1
```

```python
page_count.to_csv('./data/page_count.csv')
```

```python
page_count.head()
```

```python
student_raw = pd.read_csv('./data/aggregatelogdata.csv')
student_raw = student_raw.set_index('id_student')
```
